```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
model=Sequential()

model.add(Dense(16,input_shape=(10,),activation='relu'))

model.add(Dense(32,activation='relu'))

model.add(Dense(64,activation='relu'))

model.add(Dense(128,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_accuracy',mode='min',verbose=1,patience=25)
```

```python
y_train1=np.array(y_train)
y_test1=np.array(y_test)
```

```python
model.fit(X_train,y_train1,epochs=500,validation_data=(X_test,y_test1),callbacks=[early_stop])
```

```python
pred=model.predict_classes(X_test)
```

```python
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```
